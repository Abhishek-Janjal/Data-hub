# Using Matplotlib and Seaborn for EDA
#### In EDA, these libraries help uncover patterns, trends, and anomalies through visual representation

# Data Preparation: Load and clean your dataset using pandas.

```python
import pandas as pd
data = pd.read_csv('https://raw.githubusercontent.com/Abduttayyeb/Visual-Data-Analysis/master/Graph_Data.csv')
data.head(3)
```

# Univariate Analysis: Analyze individual variables to understand their distribution.

## Histogram with Matplotlib:

```python
import matplotlib.pyplot as plt
plt.hist(data['Age'], bins=10)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Age')
plt.show()
```

## KDE Plot with Seaborn:

```python
import seaborn as sns
sns.kdeplot(data['Age'], fill=True)
plt.title('KDE Plot of Age')
plt.show()
```

## Boxplots with Seaborn

```python
sns.boxplot(data['Age']);
```

# Bivariate Analysis: Examine relationships between two variables.

## Scatter Plot with Matplotlib:

```python
plt.figure(figsize=(7,7))
plt.scatter(data['Age'], data['Fare'])
plt.xlabel('Age')
plt.ylabel('Fare')
plt.title('Scatter Plot between Age and Fare')
plt.show()
```

## Joint Plot with Seaborn:

```python
sns.jointplot(x='Age', y='Fare', data=data, kind='scatter')
plt.figure(figsize=(12,12))
plt.show()
```

# Multivariate Analysis: Explore interactions among multiple variables.

## Pair Plot with Seaborn:

```python
sns.pairplot(data)
plt.show()
```

## Heatmap of Correlation Matrix:

```python
from sklearn.preprocessing import OrdinalEncoder
data[data.select_dtypes(include='object').columns]=OrdinalEncoder().fit_transform(data[data.select_dtypes(include='object').columns])
correlation = data.corr()
sns.heatmap(correlation, annot=True,)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

# some More Plots

### Bar Plot

```python
plt.figure(figsize = (10,10))
sns.countplot(data['Gender'], label = 'Count')
```

## Pie plot

```python
values = [20, 55, 5, 17, 3]
colors = ['g', 'r', 'y', 'b', 'm']
labels = ["Apple", "Google", "T", "TSLA", "AMZN"]
explode = [0, 0.2, 0, 0, 0.2]
# Use matplotlib to plot a pie chart 
plt.figure(figsize = (10, 10))
plt.pie(values, colors = colors, labels = labels, explode = explode)
plt.title('Stock Portfolio')
```
